```python
def plot_graphs(column_name):
    sns.displot(kind='hist',data=df,x=column_name,label=column_name,kde=True)
    sns.catplot(kind='box', data=df,x=column_name)
```

```python
num_columns = df.select_dtypes(include=['float64','int64']).iloc[:,[3,4,6,9,13,13,16]].columns
```

```python
for col in num_columns:
    plot_graphs(col)
```
